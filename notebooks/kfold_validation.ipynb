First we will hold out 15 percent of the data for the test set and we will do Stratified K-Fold Validation with 5 splits.

```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold
df = pd.read_csv(r"C:\Users\MANAAR HAQUE\Desktop\University\6th Sem\Deep Learning for Medical Image Processing\Alzheimer Detection\outputs\master_manifest.csv")

# 1. Get unique patients
unique_patients = df[['patient_id', 'label']].drop_duplicates()

# 2. STEP ONE: Hold out the Test Set (25 patients)
train_val_patients, test_patients = train_test_split(
    unique_patients, 
    test_size=0.15, 
    stratify=unique_patients['label'], 
    random_state=42
)

test_slices = df[df['patient_id'].isin(test_patients['patient_id'])]  

# 3. STEP TWO: Prepare K-Fold for the remaining 137 patients
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print(f"Total Patients: {len(unique_patients)}")
print(f"--- TEST SET (Hold-out): {len(test_patients)} patients ---")
print(f"--- K-FOLD SET (Train/Val): {len(train_val_patients)} patients ---")

p_ids = train_val_patients['patient_id'].values
p_labels = train_val_patients['label'].values

for fold, (t_idx, v_idx) in enumerate(skf.split(p_ids, p_labels)):
    # Training patients for this fold
    fold_train_ids = p_ids[t_idx]
    # Validation patients for this fold
    fold_val_ids = p_ids[v_idx]
    
    print(f"Fold {fold+1}: {len(fold_train_ids)} Train pts, {len(fold_val_ids)} Val pts")
```

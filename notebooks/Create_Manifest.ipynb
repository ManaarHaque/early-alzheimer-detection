Create a lookup table connecting every single image file to a specific patient id

```python
import pandas as pd
import os

# Update this to your 2D slices folder
slice_dir = r"C:\Users\MANAAR HAQUE\Desktop\University\6th Sem\Deep Learning for Medical Image Processing\Alzheimer Detection\outputs\2D_Slices" 

data = []

# Loop through the folders we already made
for diag in ['AD', 'MCI', 'CN']:
    folder = os.path.join(slice_dir, diag)
    if not os.path.exists(folder):
        continue
        
    for filename in os.listdir(folder):
        if filename.endswith('.png'):
            # Filenames look like: 002_S_0729_MPRAGE_..._s60.png
            # We extract '002_S_0729'
            parts = filename.split('_')
            patient_id = f"{parts[0]}_{parts[1]}_{parts[2]}" 
            
            data.append({
                'slice_path': os.path.join(diag, filename), 
                'patient_id': patient_id, 
                'label': diag
            })

df = pd.DataFrame(data)
output_csv_path = r"C:\Users\MANAAR HAQUE\Desktop\University\6th Sem\Deep Learning for Medical Image Processing\Alzheimer Detection\outputs\master_manifest.csv"
df.to_csv(output_csv_path, index=False)
print(f"Success! Manifest created with {len(df)} slices from {df['patient_id'].nunique()} patients.")
```
